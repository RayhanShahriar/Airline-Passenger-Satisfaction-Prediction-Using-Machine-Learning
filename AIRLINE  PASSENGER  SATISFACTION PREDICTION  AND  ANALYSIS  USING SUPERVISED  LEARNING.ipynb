```python
df1 = df.drop_duplicates()
```

```python
df2 = df1
df2['arrival_delay_in_minutes'] = df1['arrival_delay_in_minutes'].fillna(df1['arrival_delay_in_minutes'].median())
df2.isnull().sum()
```

```python
import seaborn as sns
```

```python
print(df2.columns.tolist())
```

```python
cols_to_scale = ['Unnamed: 0','age','flight_distance', 'inflight_wifi_service', 'departure_arrival_time_convenient', 'ease_of_online_booking', 'gate_location', 'food_and_drink', 'online_boarding', 'seat_comfort', 'inflight_entertainment', 'onboard_service', 'leg_room_service', 'baggage_handling', 'checkin_service', 'inflight_service', 'cleanliness','departure_delay_in_minutes','arrival_delay_in_minutes']

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df2[cols_to_scale] = scaler.fit_transform(df2[cols_to_scale])
```

```python
df2
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder() # instantiate label encoder in an object le
df2['Gender' ] =le.fit_transform(df2['Gender']) 
df2['customer_type' ] =le.fit_transform(df2['customer_type']) 
df2['type_of_travel' ] =le.fit_transform(df2['type_of_travel']) 


```

```python
df2
```

```python
df2 = pd.get_dummies(data=df2, columns=['customer_class'], dtype= int)
```

```python
df2
```

```python
from sklearn.model_selection import train_test_split
y = df2['satisfaction']
X = df2.drop(columns = 'satisfaction')
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 20, stratify = y)
```

```python
X_train
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, accuracy_score
knn_classifier = KNeighborsClassifier() 
knn_classifier.fit(X_train, y_train)
y_pred = knn_classifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix

confus_matrix = confusion_matrix(y_test, y_pred)
CM = sns.heatmap(confus_matrix, annot=True, fmt='d', )
CM.set(title='Confusion Matrix for the Airline Passenger Satisfaction Prediction Model')
CM.set_xlabel("Actual Satisfaction")
CM.set_ylabel("Predicted Satisfaction")
```

```python
print(classification_report(y_test, y_pred))
print("Accuracy Score {}".format(accuracy_score(y_test,y_pred)))
print("Train Accuracy: ", accuracy_score(y_train, knn_classifier.predict(X_train)))
print("Test Accuracy: ", accuracy_score(y_test, y_pred))
```

```python
import numpy as np
n_neighbours = [int(x) for x in np.linspace(start = 17000, stop = 20000, num = 5)]


grid_params = { 'n_neighbors' : n_neighbours,
               'weights' : ['uniform','distance'],
               'metric' : ['minkowski','euclidean','manhattan'],
                
              }

print(grid_params)
```

```python
from sklearn.model_selection import RandomizedSearchCV
knnclf=KNeighborsClassifier()

knnclf_randomcv=RandomizedSearchCV(estimator=knnclf,param_distributions=grid_params,n_iter=20,cv=3,verbose=2,
                               random_state=100,n_jobs=1)
```

```python
knnclf_randomcv.fit(X_train,y_train)
```

```python
from sklearn.model_selection import GridSearchCV
knnclf=KNeighborsClassifier()
grid_search=GridSearchCV(estimator=knnclf,param_grid=grid_params,cv=10,n_jobs=-1,verbose=2)
```

```python
grid_search.fit(X_train,y_train)
```

```python
grid_search.best_estimator_
```

```python
best_grid=grid_search.best_estimator_
```

```python
best_grid
```

```python
y_pred=best_grid.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix

confus_matrix = confusion_matrix(y_test, y_pred)
CM = sns.heatmap(confus_matrix, annot=True, fmt='d', )
CM.set(title='Confusion Matrix for the Airline Passenger Satisfaction Prediction Model')
CM.set_xlabel("Actual Satisfaction")
CM.set_ylabel("Predicted Satisfaction")
```

```python

print("Accuracy Score {}".format(accuracy_score(y_test,y_pred)))
print("Classification report: {}".format(classification_report(y_test,y_pred)))
```
