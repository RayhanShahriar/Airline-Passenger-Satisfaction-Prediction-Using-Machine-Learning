```python
def findOutliers(dataset):
    outliers = []
    Q2 = dataset.median()
    Q1 = dataset.describe()[4]
    Q3 = dataset.describe()[6]
    
  
    IQR = Q3 - Q1
    lf = Q1 - 1.5*IQR 
    uf = Q3 + 1.5*IQR 
    for x in dataset:
        if x < lf or x > uf:
            outliers.append(x)
    return outliers
```
